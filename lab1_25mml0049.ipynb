```python
from bs4 import BeautifulSoup
import re
```

```python
file_path="/Users/shreya/Downloads/pg1513-h/pg1513-images.html"
```

```python


with open(file_path, "r", encoding="utf-8") as f:
    html = f.read()

soup = BeautifulSoup(html, "html.parser")


paragraphs = [p.get_text(" ", strip=True) for p in soup.find_all("p")]


bad_keywords = [
    "Project Gutenberg", "License", "Title:", "Author:", 
    "ebook", "www.gutenberg.org", "End of the Project Gutenberg"
]

clean_paragraphs = []
for p in paragraphs:
    if len(p) < 20:         
        continue
    if any(k in p for k in bad_keywords):
        continue
    clean_paragraphs.append(p)

story_text = " ".join(clean_paragraphs)


story_text = re.sub(r"\s+", " ", story_text).strip()

print(story_text[:2000])


```

```python
import nltk

nltk.download("punkt")
nltk.download("averaged_perceptron_tagger")
```

```python

from nltk.tokenize import sent_tokenize, word_tokenize

sentences = sent_tokenize(story_text)
print("SENTENCES:")
print(sentences[:5])    

words = word_tokenize(story_text)
print("\nWORDS:")
print(words[:20])        

pos_tags = nltk.pos_tag(words)
print("\nPOS TAGGING:")
print(pos_tags[:20])     
```
